```python
import pandas as pd
import numpy as np

df = pd.read_csv("Data/Zip_Summary.csv")
df.head()
```

```python
# Assuming your DataFrame is called 'df'
df['num_zip'] = df['num_zip'].astype(str).str.zfill(5)
df
```

```python
# Get top 10 worst states 

worst_states = df[['num_state' , 'num_zip' , 'mu_eui']]
worst_states = worst_states.groupby(by=['num_state']).mean('mu_eui').reset_index().sort_values(by=['mu_eui'] , ascending = False).reset_index()[0:10][['num_state' , 'mu_eui']]
worst_states['num_state'].values
```

```python
# Create a list of dictionaries for each state
formatted_data = []
for _, row in worst_states.iterrows():
    state_dict = {
        'name': row['num_state'],
        'y': row['mu_eui'],
        'drilldown': row['num_state']
    }
    formatted_data.append(state_dict)

# Print the formatted data
for data in formatted_data:
    print(data)
```

```python
df_bad_states = df[df['num_state'].isin(['VT', 'ME', 'NH', 'RI', 'PA', 'MA', 'CT', 'NY', 'MN', 'TN'])][['num_state'  ,'num_zip' , 'mu_eui']]
df_bad_states
```

```python
# Sort the DataFrame by 'mu_eui' column in descending order
sorted_data = df_bad_states.sort_values('mu_eui', ascending=False)

# Group the sorted data by 'num_state'
grouped_data = sorted_data.groupby('num_state')

# Create a list of dictionaries for each state, including only the top 10 zip codes
formatted_data = []
for state, group in grouped_data:
    top_10_zip = group.head(10)  # Select only the top 10 zip codes with the highest EUI
    # Convert the zip codes to integers using the astype() method
    top_10_zip['num_zip'] = top_10_zip['num_zip'].astype(str)
    state_dict = {
        'name': state,
        'id': state,
        'data': top_10_zip[['num_zip', 'mu_eui']].values.tolist()
    }
    formatted_data.append(state_dict)

# Print the formatted data
for data in formatted_data:
    print(data)
```
