```python
import pandas as pd
import numpy as np

df = pd.read_csv("Data/County_Summary.csv")
df.head()
```

```python
# Keep relevant columns 

df_eui = df[['num_state', 'num_county', 'mu_eui']]
```

```python
# Clean the column num_county

df_eui['num_county'] = df_eui['num_county'].str.replace('_', ' ')

df_eui['num_county'] = df_eui['num_county'].str.replace('_', ' ') + ' County'
```

```python
df_eui['name'] = df_eui['num_county'] + ", " + df_eui["num_state"]
```

```python
df_eui
```

```python
# Normalize the data

import pandas as pd

# Assuming df_eui is your DataFrame containing the columns num_state, num_county, mu_eui
# Replace df_eui with the actual data or DataFrame

# Calculate the mean and standard deviation of mu_eui for each num_state
df_grouped = df_eui.groupby('num_state')['mu_eui'].agg(['mean', 'std']).reset_index()

# Merge the grouped statistics back to the original DataFrame
df_eui = pd.merge(df_eui, df_grouped, on='num_state', how='left')

# Calculate the normalized mu_eui values based on the mean and standard deviation of each num_state
df_eui['norm_mu_eui'] = (df_eui['mu_eui'] - df_eui['mean']) / df_eui['std']

# Print the updated DataFrame with the new normalized column
df_eui
```

```python
# Drop NA

df_eui = df_eui.dropna()
```

```python
df_eui_grouped_state = np.round(df_eui.groupby(by=['num_state']).mean('norm_mu_eui') , 0).astype('int').reset_index()
# Create a new column "state_key"
df_eui_grouped_state['state_key'] = 'us_' + df_eui_grouped_state['num_state'].str.lower()

# Keep only 'state_key' and 'norm_mu_eui' columns
df_filtered = df_eui_grouped_state[['state_key', 'norm_mu_eui']]

# Create a dictionary using the two columns
output_dict = dict(zip(df_filtered['state_key'], df_filtered['norm_mu_eui']))

output_dict
```

```python
df_map = pd.read_json('Data/MapData.json')

df_map
```

```python
# Assuming 'df_eui' and 'df_map' are your DataFrames
merged_df = pd.merge(df_eui, df_map, on='name', how='left')

merged_df.head(10)
```

```python
merged_df = merged_df.dropna(subset=['code'])
```

```python
merged_df = merged_df.drop(columns=['value']) 
```

```python
merged_df = merged_df.rename(columns={"norm_mu_eui": "value"})
```

```python
merged_df = merged_df[['code' ,  'value']]
```

```python
merged_df
```

```python
# Initialize an empty dictionary
output_dict = {}

# Iterate over the rows of the DataFrame
for index, row in merged_df.iterrows():
    # Extract the code and value from each row
    code = row['code']
    value = int(float(row['value']))  # Convert the value to an integer
    
    # Extract the desired format from the code
    code_parts = code.split('-')
    new_code = f"{code_parts[0]}-{code_parts[1]}-{code_parts[2]}"
    
    # Add the code and value to the dictionary
    output_dict[new_code] = value
```

```python
import json
# Specify the file path for the JSON file
file_path = 'output.json'

# Dump the output_dict into the JSON file
with open(file_path, 'w') as json_file:
    json.dump(output_dict, json_file)
```
