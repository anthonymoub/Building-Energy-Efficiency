```python
import pandas as pd
import numpy as np

df = pd.read_csv("state_heat_fuel_sankey.csv")
df 
```

```python
grouped_df = df.groupby(by=['heating_fuel_type' , 'heating_system_type']).sum('total_site_energy').reset_index()
```

```python
# Function to convert scientific notation to standard numbers
def convert_scientific_notation(value):
    return '{:.2f}'.format(float(value))

# Apply the conversion function to the 'total_site_energy' column
grouped_df['total_site_energy'] = grouped_df['total_site_energy'].apply(lambda x: convert_scientific_notation(x))
grouped_df = grouped_df[['heating_fuel_type' , 'heating_system_type' , 'total_site_energy']]
grouped_df['heating_system_type'] = grouped_df['heating_system_type'].str.title()
grouped_df
```

```python
# Function to convert total_site_energy to labels
grouped_df['total_site_energy'] = grouped_df['total_site_energy'].astype('float')

def convert_to_labels(value):
    million_value = value / 1000000
    return '{:.2f} M'.format(million_value)

# Create the 'labels' column using the 'total_site_energy' column values
grouped_df['labels'] = grouped_df['total_site_energy'].apply(lambda x: convert_to_labels(x))

grouped_df
```

```python
# Convert to HighChartJS structure 

array_of_arrays = grouped_df[['heating_fuel_type', 'heating_system_type', 'total_site_energy' , 'labels']].values
array_of_arrays[:, 2] = array_of_arrays[:, 2].astype(float)
array_of_arrays
```
